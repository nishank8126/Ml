```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('tc dataset.csv')
print(data.head(40))
```

```python
num_col='External'
data_num=data[num_col]
mean=data_num.mean()
median=data_num.median()
mode=data_num.mode()[0]
std=data_num.std()
var=data_num.var()
range_val=data_num.max()-data_num.min()
print('statistics:')
print(f'mean={mean}')
print(f'median={median}')
print(f'mode={mode}')
print(f'std_dev={std}')
print(f'variance={var}')
print(f'range={range_val}')
```

```python
plt.figure(figsize=(8,5))
plt.hist(data_num,bins=10,color='skyblue',edgecolor='black')
plt.title(f'histogram of {num_col}')
plt.xlabel(num_col)
plt.ylabel("frequency")
plt.show()
```

```python

q1=data_num.quantile(0.25)
q3=data_num.quantile(0.75)
iqr=q3-q1
low_bound=q1-1.5*iqr
up_bound=q3+1.5*iqr
outliers=data_num[(data_num<low_bound)|(data_num>up_bound)]
print(outliers)
```

```python
def result(External):
    if External>18:
        return 'pass'
    else:
        return 'fail'
data['res']=data['External'].apply(result)
cat_col='res'
data_cat=data[cat_col]
cat_counts=data_cat.value_counts()
print("category frequencies:")
print(cat_counts)
```

```python
plt.figure(figsize=(8,5))
cat_counts.plot(kind='bar',color='coral', edgecolor='black')
plt.title(f'bar chart for {cat_counts}')
plt.xlabel(cat_col)
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
cat_counts.plot(kind='pie',autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title(f'pie chart for {cat_counts}')
plt.show()
```
